```python


text1 = """
Often referred to as "King Kohli," Virat Kohli is redefined modern batting through sheer consistency and an unrelenting elite fitness standard. Known for his aggressive passion and technical mastery, he is considered one of the greatest "chasers" in the history of Limited Overs Cricket. With over 80 international centuries, Kohli transitioned from a young prodigy to a global sporting icon, leading the Indian Test team to historic victories abroad and instilling a culture of "pace-bowling dominance" and world-class athleticism.


"""

text2 = """
Known as "Captain Cool," Mahendra Singh Dhoni is celebrated as one of the most tactical minds to ever play the game. He is the only captain in history to win all three major ICC trophies: the T20 World Cup (2007), the ODI World Cup (2011), and the Champions Trophy (2013). Beyond his leadership, Dhoni was a lightning-fast wicketkeeper and a legendary finisher who could remain calm under immense pressure, famously ending the 2011 World Cup final with his signature six.
"""

text3 = """
Nicknamed "The Hitman," Rohit Sharma is synonymous with effortless elegance and massive scores. He holds the world record for the highest individual score in an ODI (264) and is the only player to have hit three double-centuries in the format. As a leader, he is known for his "selfless" batting approach and tactical brilliance, culminating in leading India to the 2024 ICC T20 World Cup title. His signature pull shot is widely regarded as one of the most beautiful sights in cricket.
"""

text4 = """
Often called "Sir Jadeja" by fans and teammates, he is the ultimate "3D" player—excelling in batting, bowling, and fielding. As a left-arm spinner, he is a nightmare for batsmen on turning tracks, and as a lower-order batter, he has evolved into a reliable match-winner. However, it is his fielding that truly sets him apart; with a "rocket arm" and incredible reflexes, he is arguably the greatest Indian fielder of his generation. He remains a vital pillar of the Indian team across all formats.
"""



from langchain_core.documents import Document


doc1 = Document(
    page_content=text1,
    metadata = {'player': 'Virat Kholi'}
)

doc2 = Document(
    page_content=text2,
    metadata = {'player': 'Mahendra Singh Dhoni'}
)

doc3 = Document(
    page_content=text3,
    metadata = {'player': 'Rohit Sharma'}
)

doc4 = Document(
    page_content=text4,
    metadata = {'player': 'Ravindra Jadeja'}
)

docs = [doc1, doc2, doc3, doc4]
```

```python
pip install langchain-chroma
```

```python
from langchain_chroma import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_core.documents import Document
from google.colab import userdata

api_key = userdata.get('GEMINI_API_KEY')

embeddings = GoogleGenerativeAIEmbeddings(model="gemini-embedding-001", api_key=api_key)

vector_store = Chroma.from_documents(
    documents=docs,
    embedding=embeddings,
    collection_name="cricket_players"
)
```

```python
retriever = vector_store.as_retriever(search_kwargs={"k": 2})
```

```python
query2 = "Who is Virat Kholi"
results = retriever.invoke(query2)
# print(results[0].page_content)

for res in results:
  print("--------------------------")
  print(res.page_content)
  print("--------------------------")
```

```python
from langchain_community.vectorstores import FAISS

vector_Store = FAISS.from_documents(
    documents=docs2,
    embedding=embeddings
)
```

```python
retriever = vector_Store.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 3, "lambda_mult": 1}
)
```

```python
query2 = "What is quantum entanglement?"
result2 = retriever.invoke(query2)
```
