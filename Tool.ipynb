```python
from langchain_google_genai import ChatGoogleGenerativeAI, GoogleGenerativeAIEmbeddings
from langchain_core.tools import tool

@tool
def add(a: int, b: int) -> int:
  """ Given 2 numbers a and b, add these 2 numbers and return the result """

  return a + b

@tool
def multiply(a: int, b: int) -> int:
  """ Given 2 numbers a and b, multiply these 2 numbers and return the reesult """

  return a * b



```

```python
from google.colab import userdata
api_key = userdata.get('GEMINI_API_KEY')
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    api_key=api_key,
    temperature=0.1
)
```

```python
llm_tool = llm.bind_tools([multiply, add])
```

```python
llm_tool.invoke("How are you?")
```

```python
from langchain_core.messages import HumanMessage
query = HumanMessage("What 50 into 64")
```

```python
messages = [query]
print(messages)
```

```python


result = llm_tool.invoke(messages)
```

```python
result
```

```python
messages.append(result)
print(messages)
```

```python
tool_res = multiply.invoke(result.tool_calls[0])
```

```python
messages.append(tool_res)
print(messages)
```

```python
answer = llm_tool.invoke(messages).content
print(answer)
```

# **Currency Conversion Tool**

```python
from google.colab import userdata
currency_api = userdata.get('CURRENCY_API')
```

```python
import requests

@tool
def conversion(from_curr: str, to_curr: str, amount: int) -> int:
  """ Use this tool to  convert from one currency to another as it will have real time conversion data"""
  url = f'https://v6.exchangerate-api.com/v6/{currency_api}/pair/{from_curr}/{to_curr}/{amount}'
  response = requests.get(url)
  data = response.json()
  return data["conversion_result"]
```

```python
llm_tool = llm.bind_tools([multiply, add, conversion])
```

```python
curr_query = HumanMessage("Convert 500 indian rupees to australian dollars")
curr_messages = [curr_query]
```

```python
res = llm_tool.invoke(curr_messages)
curr_messages.append(res)
```

```python
res
```

```python
print(curr_messages)
```

```python
tool_ans = conversion.invoke(res.tool_calls[0])
curr_messages.append(tool_ans)
```

```python

print(curr_messages)
```

```python
final_result = llm_tool.invoke(curr_messages).content

print(final_result)
```
